```
#finding the population percentage of each continent
cont_data_percentage = data.groupby('Continent')['World Population Percentage'].sum().round(1).sort_values(ascending= False).reset_index()
cont_data_percentage
```

```
labels= cont_data_percentage['Continent']
labels
```

```
pop_percent = cont_data_percentage['World Population Percentage']
pop_percent
```

```
pop_percent.plot.pie(labels = labels, autopct = '%1.2f%%', figsize= (10,7),shadow = True)
plt.title('Continent - Population%')
```

**Moreover when we compare the continent with world population percentage, Asia occupies 59.2%**

# COMPARING THE CONTINENT POPULATION OF 1970 AND 2022

```
ppl_diff_70to22 = data.groupby('Continent')[['1970 Population','2022 Population']].sum().sort_values(by = 'Continent', ascending= True).reset_index()
ppl_diff_70to22
```

```
ppl_diff_70to22['Percentage'] = round(ppl_diff_70to22['1970 Population']/ ppl_diff_70to22['2022 Population']*100,2)
ppl_diff_70to22
```

# TOP 5 COUNTRIES FROM THE CONTINENT

```
def top_5_countries(Continent):
    cont = data[data['Continent'] == Continent]
    return cont.groupby('Country')['2022 Population'].sum().sort_values(ascending = False).head(5)
```

# BOTTOM 5 COUNTRIES FROM THE CONTINENT

```
def bottom_5_countries(Continent):
    cont = data[data['Continent'] == Continent]
    return cont.groupby('Country')['2020 Population'].sum().sort_values(ascending = True).head(5)
```

# COUNTRY VS AREA

```
population_2022=data.sort_values(by="Area (km²)",ascending=False).head(10)
fig, ax = plt.subplots(figsize =(10,7))
x=population_2022["Country"]
y=population_2022["Area (km²)"]
plt.pie(y,labels=x,autopct='%1.1f%%')
plt.show()
```

**Russia is the top countries according to the area which occupies 23.2%**

# COUNTRY VS WORLD POPULATION PERCENTAGE

```
population_2022=data.sort_values(by="World Population Percentage",ascending=False).head(10)
fig, ax = plt.subplots(figsize =(10,7))
x=population_2022["Country"]
y=population_2022["World Population Percentage"]
plt.pie(y,labels=x,autopct='%1.1f%%')
plt.show()
```
